```python
import pandas as pd
import seaborn as sns
data=pd.read_excel("Karachi_car_data.xlsx")
data
```

```python
data.drop(['Ad Id','URL','City','itemCondition','color','priceCurrency', 'car_specs','Brand','Name'],axis=1, inplace=True)
data

```

```python
data.isna().sum()

```

```python
data.dropna(inplace=True)
```

```python
column_mapping = {
    'fuelType': 'fuel_type',
    'vehicleTransmission':'transmission',
    'mileageFromOdometer (km)': 'kms_driven',
    'Offering_Price': 'price_pkr',
    'Engine_Displacement(cc)': 'engine_cc',
    'modelDate': 'model_year',
    'vehicle_transmission': 'transmission',
    'Model': 'car_name',
    
}

data.rename(columns=column_mapping, inplace=True)

data
```

# dropping duplicates

```python
data= data[~data.duplicated(subset=['car_name','price_pkr', 'model_year', 'kms_driven', 'transmission', 'fuel_type','engine_cc','manufacturer' ])]
data
```

# Two Words brands Exceptions >>>>: 'Mercedes Benz', 'Range Rover','Land Rover','Alfa Romeo','JW Forland','Classic Cars'

```python
exceptions=['Mercedes Benz', 'Range Rover','Land Rover','JW Forland']
corrected= ['Mercedes-Benz', 'Range-Rover','Land-Rover','JW-Forland']
for i in range(len(exceptions)):
    data= data.replace(exceptions[i], corrected[i])
    
```

# Correlation Check

```python
def correlation (dataset, threshold):
    col_corr=set()
    corr_matrix=data.corr()
    
    for i in range (len(corr_matrix.columns)):
        for j in range (i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
corr_features=correlation(data, 0.7)
corr_features
```
